```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first='True')
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependent=dataset[['charges']]
```

```python
#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':['rbf','poly','sigmoid','linear'],
           'C':[10,100,1000,2000,3000],
           'gamma':['auto','scale']}
#regressor=SVR(kernel="poly",degree=4,gamma="auto",coef0=1,epsilon=0.5,cache_size=500,verbose=True,max_iter=1)
#regressor=SVR(kernel="poly")
#regressor.fit(X_train,y_train)

#table
```

```python
grid = GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=1)
grid.fit(independent,dependent)
re=grid.cv_results_
```

```python
print("The r_score value for best paramter{}:".format(grid.best_params_))
table=pd.DataFrame(re)
table
```
